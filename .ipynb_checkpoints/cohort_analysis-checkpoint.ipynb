point_event_logs: 포인트 지급 로그

```python
point_event_logs = pd.read_csv("task_5_dataset/point_event_logs.csv", parse_dates=['event_timestamp'])
point_event_logs.tail(3)
```

order_logs : 유저의 주문 정보(하나의 주문에 하나의 레코드 기록)
- 주문이 발생한 시간과 user_id 정보 확인
- 주문 데이터 기간은 `2019-06-03 ~ 2019-06-30`
- column 설명 
    - order_timestamp : 주문이 발생한 시간(한국 시간 기준)
    - user_id : 이용자 id
    - order_number : 주문 번호 
        - 쇼핑몰을 통합하여 한번에 주문하는 주문 구분 단위
    - total_item_amount : 주문 상품의 총 금액
        - 주문한 모든 상품 값을 합산한 금액
    - total_shipping_fee : 주문의 총 배송비
        - 주문한 모든 배송비를 합산한 금액
    - total_point_spent : 주문에 사용된 총포인트
        - 상품의 금액 + 배송비에서 포인트를 사용하여 차감되는 금액

```python
order_logs = pd.read_csv("task_5_dataset/order_logs.csv", parse_dates=['order_timestamp'])
order_logs.tail(3)
```

### 실험군(포인트 지급 받은 대상)과 대조군(포인트 미지급 대상)의 일자별 총 매출

```python
target_logs = point_event_logs.copy()
target_logs["date"] = target_logs["event_timestamp"].apply(lambda x: str(x)[0:10])
target_logs = target_logs.loc[(target_logs['date'] == "2019-06-10") & (target_logs['event_type'] == "point_add")]
target_logs['target'] = '실험군'
target = target_logs[['user_id', 'target']]

orders = pd.merge(order_logs, target, on='user_id', how='left')
orders.fillna('대조군', inplace=True)
orders['date'] = orders['order_timestamp'].apply(lambda x: str(x)[:10])
orders = orders[['date', 'target', 'total_item_amount']]

revenues = orders.groupby(['date', 'target'])['total_item_amount'] \
                 .agg({'buyers': 'count', 'total_revenue': 'sum', 'amount_per_person': 'mean'}) \
                 .reset_index()
revenues['date'] = revenues['date'].apply(lambda x: datetime.datetime.strptime(x, "%Y-%m-%d").date())
```

```python
plt.figure(figsize=(12, 7))
ax = sns.lineplot(x='date', y='total_revenue', hue='target', data=revenues, sort=True)
plt.title("일자별 총 매출")
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```

- Reward point가 지급된 2019-06-10와 reward point 소멸 직전의 2019-06-23사이에 매출 증대가 큰 폭으로 발생했음을 확인 할 수 있습니다.

### 실험군(포인트 지급 받은 대상)과 대조군(포인트 미지급 대상)의 일자별 1인당 평균 구매 금액

```python
plt.figure(figsize=(12, 7))
ax = sns.lineplot(x='date', y='amount_per_person', hue='target', data=revenues, sort=True)
plt.title("일자별 1인당 평균 구매 금액")
plt.show()
```

- 1인당 구매 금액을 보니 reward marketing으로 1인당 구매 금액이 늘어난 것은 아닌 것으로 짐작 할 수 있습니다.

### 이중차분법(Difference-in-Differences Estimation)란?

### 이중차분법의 전제 조건
- 실험군과 대조군은 개입 받기 전의 트렌드가 평행해야 합니다.
- 개입이 있는 동안 다른 변화는 없어야 합니다.



Reward Marketing로 인한 매출 증대가 있었는지 알아보겠습니다. \
EDA를 살펴보니 포인트 지급 받기 전의 두 그룹의 매출 트렌드가 비슷하다고 볼 수 있었으며 포인트 지급 기간에 다른 변화는 없다고 가정하였습니다. \
이중차분법을 직접 계산하는 방법과 회귀 두가지 방법을 통해 살펴보겠습니다.

### Difference in Difference by Regression 

$y = \beta_0 + \beta_1D^{Post} + \beta_2D^{Tr} + \beta_3D^{Post}D^{Tr} [+ \beta_4X] + \epsilon$

<div>
<img src="img/ddregressiontable.png" width="500"/> 
</div>

회귀를 통한 이분차분법의 이점은 다음과 같습니다.
- 부트스트래핑처럼 많은 계산을 하지 않아도 빠르고 안정적으로 모형 계수의 표준오차를 알 수 있기 때문에 계수의 신뢰도를 추정할 수 있습니다.
- 회귀 분석시 종속변수에 영향을 미치는 모든 독립변수를 회귀모형에 포함하지 않는 한 모형의 가중치는 항상 편향된(biased) 값이 나옵니다. formular에 conditional(ex; `[+ \beta_4X]`)을 추가할 수 있기 때문에 가중치 추정값의 편향됨(biased)을 줄여 줄 수 있습니다. 
- estimate의 정확도가 높습니다. 

## 4. 데이터 분석

```python
order_logs['date'] = order_logs['order_timestamp'].apply(lambda x: str(x)[0:10])
orders = order_logs.drop(['order_timestamp', 'total_shipping_fee', 'total_point_spent', 'order_number'], axis=1)
orders = orders[orders['date'] < '2019-06-24']

points = point_event_logs[point_event_logs['event_type']=='point_add'][['user_id', 'amount']]
points['treatment'] = points['amount'].apply(lambda x: 1 if x > 0 else 0)

orders_tr = orders.merge(points[['user_id', 'treatment']], on='user_id', how='left').fillna(0).drop(['user_id'], axis=1)
orders_tr = orders_tr.groupby(['date', 'treatment']).agg('sum').reset_index()
orders_tr['post'] = orders_tr['date'].apply(lambda x: 1 if x >= '2019-06-10' else 0)
```

### 이중차분법을 직접 계산

```python
orders_tr.groupby(['treatment', 'post']).agg('mean')
dd = (4.334861e+06 - 4.300829e+05) - (1.061500e+05 - 5.353333e+04)
print("{:,}".format(int(dd)))
```

### 이중차분법을 회귀로 계산

```python
model = sm.OLS.from_formula('total_item_amount ~ post + treatment + post:treatment', orders_tr)
result = model.fit()
print(result.summary())
```

```python
# post:treatment coefficience
print("{:,}".format(int(3.852e+06)))
```

## 5.결과 
Regression과 직접 계산한 결과값이 유사함을 확인할 수 있었습니다. \
Reward marketing으로 3,852,000원의 매출 증대가 발생 했음을 추론 할 수 있습니다.
